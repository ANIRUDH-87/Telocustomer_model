Preparing data in a form that a machine learning model can understand by Feature Engineering.

```python
y = df["Churn"] # It is a target variable what we need to predict in model
```

```python
x =df.drop("Churn", axis = 1) #Removing target columns from inputs
```

If we didn't drop this column (or) Churn will stays in X Model Cheats and gives fake Accuracy.

```python
x = x.drop("customerID", axis = 1) # Removing ID Column
```

Removing the Customer ID column because it doesn't tells why customer churns. It just a unique label not behaviour

```python
#Now we are converting target variable into numbers

y = y.map({"Yes": 1, "No": 0}) # Converting to Binary form
```

Why Because ML Cannot learn from text. Binary classification needs binary values

```python
#Now we will identify which columns are in text
categorical_cols = x.select_dtypes(include = ["object"]).columns
```

```python
categorical_cols
```

```python
#Now we will  do One - hot Encoding.To convert  catogorical to numerical columns
x =pd.get_dummies(x, columns = categorical_cols, drop_first = True)

#pd.get_dummies() -> converts text into binary (0/1)
#Now all columns in 'x' are numeric
```

Because to avoid redudancy, Multicollinearity and Improve model stability.

```python
#Now for better feature understanding we divide Tenure into buckets

x["tenure_group"] = pd.cut(
    x["tenure"],             #This Column we are grouping
    bins = [0, 12, 48, 72],  #Bucket boundaries
    labels = ["New", "Mid", "Long"]  #Human readable group names
)
```

This ranges are business friendly and commonly used.

```python
# Now we Find the median of MonthlyCharges
median_charge = x["MonthlyCharges"].median()

# Creating a binary feature:
# 1 = High charges
# 0 = Low charges
x["High_Monthly_Charges"] = (x["MonthlyCharges"] > median_charge).astype(int)
```

```python
# Encode the newly created tenure_group column
X = pd.get_dummies(x, columns=["tenure_group"], drop_first=True)
```

Now in this Feature engineering we added more features which i svery useful for ML model.

Now we will move to Train -Test Split

```python
# function to split data into train & test

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,                      # input features
    y,                      # target variable
    test_size=0.2,          # 20% data for testing
    random_state=42,        # ensures reproducibility
    stratify=y              # preserves churn ratio in train & test
)
```

```python
# We using logistic regression model for binary classification

from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(max_iter=1000)
```

Why we created max_iter is if iterations ae too low model get fails and gives warnings.

```python
 # train the model using training data
log_model.fit(X_train, y_train)
```

Here .fit() tells model to learn x_train for Input features Y_train for correct answers

```python
y_pred = log_model.predict(X_test) # It is predicting churn for unseen data
```

The main point is we Never predict on training data for evaluation.we Always predict on test data.

Now we successfully bulid the model

Now we are doing main step that is Evaluating the model. To check how the model's predictions are good by comparing actual values and predicted values
